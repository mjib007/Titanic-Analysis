## 🩹 Ch3. 第 2 關：補洞 + 翻譯

### 為什麼資料需要「修整」？

在機器學習的世界裡，資料的品質決定了模型的生死。真實世界的資料**不會乖乖長成電腦喜歡的樣子**，我們通常會遇到以下兩大難關：

#### **狀況 1：資料有「破洞」（缺失值）**

對於電腦來說，遇到「空格」會直接當機或無法運算——就像老師問你「$3 + \text{空白} = ?$」一樣，因為沒有數值，邏輯就斷掉了。

#### **狀況 2：資料是文字，電腦只懂數字**
電腦的大腦本質上是大型計算機，它看到 `male`、`female` 或港口縮寫 `S`、`C`、`Q` 時會感到混亂。為了讓模型理解，我們必須將這些文字**「翻譯」成數字格式**。

---

>
> 想像你是一位 HR，收到一疊應徵者的履歷：

> * **進行翻譯**：有些是**外國語系的履歷**，你必須先找人翻譯成中文，你才能進行下一步的篩選與評分。

### 📋 接下來的程式碼會做四件事：

1.  **檢查破洞**：找出哪些欄位有遺失值，數量有多少？
2.  **補洞與捨棄**：將完全不相關的欄位丟掉，並對遺失的年齡、票價填入「平均值」。
3.  **語意翻譯**：
    * 將性別 `sex` 轉換為 $0$ 與 $1$。
    * 將港口 `embarked` 轉換為 **One-hot Encoding** 格式。
4.  **公平競爭（標準化）**：將 `age` 與 `fare` 的範圍縮小到 $0$ 至 $1$ 之間，避免模型因為數值大小而產生偏見。

````python
# === 步驟 1:檢查 X 中哪些欄位有缺失值 ===
missing_values_X = X.isnull().sum()
missing_values_X = missing_values_X[missing_values_X > 0]

if not missing_values_X.empty:
    print("X 中有遺失值的欄位及其數量:")
    print(missing_values_X)
else:
    print("X 中沒有遺失值的欄位。")

# === 步驟 2:刪除不需要的欄位、填補缺失值 ===
# 遺失值太多的就刪除,不相干的也不必要
X = X.drop(columns=["name", "ticket", "cabin", "boat", "body", "home.dest"])

# 填補 'age' 欄位的缺失值,使用平均值
X['age'] = X['age'].fillna(X['age'].mean())

# 填補 'fare' 欄位的缺失值,使用平均值
X['fare'] = X['fare'].fillna(X['fare'].mean())

# 填補 'embarked' 欄位的缺失值,使用眾數
X['embarked'] = X['embarked'].fillna(X['embarked'].mode()[0])

print("\n處理後的 X 遺失值:")
print(X.isnull().sum()[X.isnull().sum() > 0])

# === 步驟 3:把文字欄位翻譯成數字 ===
# 將 'embarked' 欄位轉換為 One-hot encoding
X = pd.get_dummies(X, columns=['embarked'], prefix='embarked')

# 將 'sex' 欄位的 'female' 轉為 0, 'male' 轉為 1
X['sex'] = X['sex'].map({'female': 0, 'male': 1})

print("\n轉換後的 X 資料集前五行:")
print(X.head())

# === 步驟 4:把 age 和 fare 拉到同一個範圍(0~1)===
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X[['age', 'fare']] = scaler.fit_transform(X[['age', 'fare']])

print("\nMin-Max 標準化後的 X 資料集前五行:")
print(X.head())
````

### 📊 執行後預期結果

完成資料預處理的程式碼後，你應該會觀察到以下變化：

1.  **缺失值消失**：
    * 原本 `age` 缺失約 $263$ 筆、`fare` 缺失 $1$ 筆、`embarked` 缺失 $2$ 筆。
    * 處理後再次檢查，輸出應為空值，代表所有洞都補齊了。
2.  **特徵轉譯完成**：
    * `sex` 欄位已由文字變為數字 $0$ 或 $1$。
    * `embarked` 欄位被拆解為 `embarked_C`、`embarked_Q`、`embarked_S` 三個欄位。
3.  **數值範圍統一**：
    * 原本範圍極大的 `age` ($0 \sim 80$) 與 `fare` ($0 \sim 500$)，現在都變成了 $0 \sim 1$ 之間的小數。

### ⚠️ 重點一：為什麼 sex 用 0/1，但 embarked 用 One-hot？

這是初學者最容易產生困惑的地方，讓我們花 30 秒釐清這兩種編碼的適用情境：

| 編碼方式 | 適用情境 | 範例 |
| :--- | :--- | :--- |
| **0/1 (Label Encoding)** | 只有**兩種選項**，且兩者沒有大小之分 | `sex` (只有 male / female) |
| **One-hot Encoding** | 有 **3 種以上選項**，且選項間沒有順序意義 | `embarked` (C / Q / S 三個港口) |

**為什麼 embarked 不能直接用 0, 1, 2？**
如果你將 C 設為 $0$、Q 設為 $1$、S 設為 $2$，模型會誤以為「**S ($2$) 比 C ($0$) 大**」，或是「**Q ($1$) 是 C 與 S 的中間值**」。但事實上，港口之間並沒有大小之分。



One-hot 的做法是將一個欄位拆成三個，確保每個港口在模型眼中都是「平起平坐」的：

| 原本港口 | embarked_C | embarked_Q | embarked_S |
| :--- | :---: | :---: | :---: |
| **C** | $1$ | $0$ | $0$ |
| **Q** | $0$ | $1$ | $0$ |
| **S** | $0$ | $0$ | $1$ |

> 💡 **生活比喻**
> * **0/1**：適合「冷 / 熱」，因為它們是互補的二元概念。
> * **One-hot**：適合「貓 / 狗 / 兔子」，因為這三種動物沒有「誰比誰大」的問題。

### ⚠️ 重點二：為什麼要 MinMax 標準化？

讓我們看一下原始資料的數值差異：
* `age` 範圍：$0 \sim 80$
* `fare` 範圍：$0 \sim 500$
* `sex` 範圍：$0 \sim 1$

如果不做標準化，模型會因為 `fare` 的數字最大，而誤以為它最重要，進而產生**「偏心」**。但歷史數據告訴我們，**性別（女性優先上救生艇）才是最關鍵的存活因素**。

MinMax 標準化能將所有特徵壓制在 $0 \sim 1$ 之間，讓所有欄位在同一個基準線上「公平上場」，模型才能客觀判斷誰才是真正的關鍵因素。

> 💡 **生活比喻**

## 🧠 Ch4. 第 3 關：蓋小腦袋

### 神經網路是什麼？

別被這個專業名詞嚇到了，**神經網路本質上就是一條「資訊加工的工廠生產線」**。

> 💡 **生活比喻：果汁工廠**
>
> * **原料入口（輸入層 Input Layer）**：
>     準備好 $9$ 種水果（對應 $9$ 個特徵，如 `pclass`、`sex`、`age`、`fare` 等）。
> * **初步加工（第 1 層隱藏層 Hidden Layer）**：
>     使用 $16$ 台果汁機，將這 $9$ 種原料進行混合攪拌，提取出第一層風味。
> * **精煉過濾（第 2 層隱藏層 Hidden Layer）**：
>     交由 $8$ 台果汁機進一步精製，過濾掉雜質，留下最關鍵的營養資訊。
> * **成品出口（輸出層 Output Layer）**：
>     最後倒出 $1$ 杯成品——也就是 AI 告訴你「這名乘客會活下來的機率」（介於 $0 \sim 1$ 之間）。
> * **品管環節（Dropout）**：
>     為了怕員工偷懶，每次工作隨機叫 $35\%$ 的工人放假，強迫剩下的人必須學會所有工序，確保工廠在任何情況下都能正常運作。



---

### 為什麼要這樣設計？（架構邏輯）

神經網路的「層數」與「神經元數量」決定了模型的學習能力：

* **層級分佈 ($9 \to 16 \to 8 \to 1$)**：
    這是一種經驗法則。對於鐵達尼號這種小型資料集，這個規模剛剛好。如果神經元過多，模型會變得太聰明進而開始「死背答案」（過擬合 Overfitting）；太少則會學不到規律。
* **設置 Dropout $0.35$**：
    訓練時隨機忽略 $35\%$ 的節點。這是一種**正則化（Regularization）**手段，強迫模型「全班都要懂」，不能只靠少數幾位資優生神經元來撐場，能有效提升模型的適應能力。
* **輸出層僅 $1$ 個節點**：
    因為我們的任務是「二元分類」（生或死），我們只需要知道一個最終數值——「生存機率」，因此輸出層只需配置一個節點。

2.  **建構模型**：依照上述「$9 \to 16 \to 8 \to 1$」的藍圖，使用 TensorFlow / Keras 蓋出這座 AI 工廠。

### 📋 接下來的程式碼會做兩件事

為了讓 AI 能夠學會預測並驗證成果，我們必須執行以下兩個核心動作：

    * **格式轉換**：將資料由 Pandas 表格轉為 NumPy 的 `ndarray` 格式，這是深度學習框架（如 TensorFlow）最喜歡的高效運算格式。

2.  **蓋出模型（Model Construction）**：
    依照我們先前規劃的藍圖，蓋出「$9 \to 16 \to 8 \to 1$」的結構。
    * **輸入層**：$9$ 個特徵維度。

    * **Dropout**：設定為 $35\%$，防止過擬合（Overfitting）。
    * **輸出層**：$1$ 維度輸出，用於二元分類（Binary Classification）。

#### 🎯 開發歷程：這段程式碼背後的指令 (Prompts)

> **指令 2：建構神經網路**
> 「請幫我建構一個 Feed Forward Neural Network：
> * `input layer` 為 $9$ 個維度的特徵向量。
> * 第一個 `hidden layer` 為 $16$ 維度。
> * 第二個 `hidden layer` 為 $8$ 維度。
> * `output layer` 為一個維度的輸出，用來做 `binary classification` 預測 survival。
> * 加入 `dropout layer` 設定 `dropout rate = 35%`。」

> 🎓 **給學生的學習小提醒**
>
> 在這個階段，我們正式將「資料」與「演算法」結合。你會看到模型摘要（Model Summary），請注意裡面的 `Param #`（參數數量），那就是模型內部可以用來學習的「小旋鈕」數量！

````python
from sklearn.model_selection import train_test_split

# 將資料切分為訓練集和測試集 (80% 訓練, 20% 測試)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 將 Pandas DataFrame/Series 轉換為 NumPy ndarray
X_train_array = X_train.to_numpy(dtype=float)
X_test_array = X_test.to_numpy(dtype=float)
Y_train_array = Y_train.to_numpy()
Y_test_array = Y_test.to_numpy()

print("X_train 陣列形狀:", X_train_array.shape)
print("Y_train 陣列形狀:", Y_train_array.shape)
print("X_test 陣列形狀:", X_test_array.shape)
print("Y_test 陣列形狀:", Y_test_array.shape)

# === 蓋模型 ===
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# 建構 FFNN 模型
model = Sequential([
    Dense(16, activation='relu', input_shape=(X_train_array.shape[1],)),  # 第一個 hidden layer
    Dropout(0.35),                                                          # Dropout layer
    Dense(8, activation='relu'),                                            # 第二個 hidden layer
    Dense(1, activation='sigmoid')                                          # Output layer for binary classification
])

# 編譯模型
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 顯示模型結構
model.summary()
````

> * **Batch size（一次背多少）= $8$**：

>     每次讀完一個章節，就從書裡抽出 $10\%$ 的題目來當「模擬考」，看看自己是不是真的掌握了重點。

### 為什麼要做模擬考 (Validation)？

這是訓練中最重要的一個細節。

因為模型非常容易陷入「**死背題目**」的陷阱——它可能記住了某個特定乘客的名字或特徵組合，而不是理解「性別」與「艙等」的通用規律。這在 AI 領域稱為 **過擬合 (Overfitting)**。

透過 **Validation（驗證集）**，我們用模型「當下沒讀過」的題目來考它：
* 如果模型「課本題目」全對，但「模擬考」全錯 $\to$ 代表它在**死背**。
* 如果模型兩者表現都很好 $\to$ 代表它真的**學會了規律**，具備處理未知資料的能力。

2.  **執行期末考與看成績 (Evaluation)**：
    * 使用那 $20\%$ 「從未看過」的測試資料來考驗模型。

````python
# 訓練模型
history = model.fit(
    X_train_array,
    Y_train_array,
    epochs=30,
    batch_size=8,
    validation_split=0.1,  # 10% 的訓練資料作為驗證集
    verbose=1              # 顯示訓練進度
)

print("\n訓練完成!")
print("訓練歷史 keys:", history.history.keys())

# === 評估模型 ===
from sklearn.metrics import classification_report, confusion_matrix

# 在測試集上進行預測
Y_pred_proba = model.predict(X_test_array)
# 將預測的機率轉換為二元類別 (0 或 1)
Y_pred = (Y_pred_proba > 0.5).astype(int)

# 輸出分類報告
print("\nClassification Report:")
print(classification_report(Y_test_array, Y_pred))

# 輸出混淆矩陣
print("\nConfusion Matrix:")
print(confusion_matrix(Y_test_array, Y_pred))
````

### ⚠️ 重點解釋：混淆矩陣 (Confusion Matrix) 怎麼看？

混淆矩陣是 AI 的**「期末考成績單」**。透過這個四宮格，我們可以清楚看到模型「在哪裡跌倒」：

| | 預測結果：死亡 | 預測結果：存活 |
| :--- | :--- | :--- |
| **真實：死亡** | ✅ **答對** (真正例 / True Negative) | ❌ **誤判為活** (偽正例 / False Positive) |
| **真實：存活** | ❌ **誤判為死** (偽負例 / False Negative) | ✅ **答對** (真正例 / True Positive) |

* **對角線（左上 & 右下）**：代表 AI 答對的人數。
* **反對角線**：代表 AI 答錯的情況（誤判）。



---

### ⚠️ 重點解釋：Precision、Recall 與 F1 分數

光看「準確率 (Accuracy)」有時會產生誤導，因為「錯的方向」在不同情境下意義不同。讓我們用醫療例子來理解：

| 指標名稱 | 白話翻譯 | 醫療應用的考量 |
| :--- | :--- | :--- |
| **Precision (精確率)** | 「我判斷有病的人中，真的有病的比例」 | 為了**避免誤診**，不想讓健康的人白做化療。 |
| **Recall (召回率)** | 「真的有病的人中，被我抓出來的比例」 | 為了**避免漏診**，不想放過任何一個潛在病人。 |
| **F1-score** | 精確率與召回率的平衡分數 | 當兩者同樣重要時，參考這個指標。 |

對鐵達尼號預測專案來說，生與死的判斷權重是相當的，因此我們主要參考 **F1-score** 或 **Accuracy** 即可。

### 🎮 玩玩看：調整參數，觀察 AI 的「學習狀態」

這是這堂課最有教育意義的環節，試著修改參數後重新執行，觀察數據變化：

* **將 `epoch` 改成 $5$**：
    AI 「讀書次數不夠」，準確率會變低。這在 AI 領域稱為 **欠擬合 (Underfitting)**。
* **將 `epoch` 改成 $300$**：
    AI 「讀得太死、鑽牛角尖」，連資料裡的雜訊都背起來了。你會發現訓練準確率極高，但「模擬考 (Val)」成績反而變差，這就是典型的 **過擬合 (Overfitting)**。
* **將 `batch_size` 改成 $64$**：
    一次背 64 題。訓練速度會變快，但模型在細微規律的掌握上可能會略遜一籌。
* **將 `dropout` 改成 $0.7$**：
    每次隨機讓 $70\%$ 的工人放假。模型學習壓力過大，可能導致訓練完全無法成形。

### 📋 接下來的程式碼會做三件事：

### 📋 接下來的程式碼會做三件事

現在我們要讓 AI 展現它的「預知能力」。為了確保預測準確，我們必須對這兩位新乘客執行與訓練資料完全一致的處理流程：

1.  **建立新乘客資料**：手動輸入 Jack 與 Rose 的各項特徵參數。
2.  **執行資料預處理**：
    * 將性別 `sex` 映射為 $0$ 與 $1$。
    * 將港口 `embarked` 展開為 One-hot Encoding 格式。
    * **關鍵點**：使用訓練階段建立的 `scaler` 對兩人進行 **MinMax 標準化**，確保數值基準與模型的大腦記憶一致。
3.  **輸出預測結果**：計算生還機率（$0 \sim 1$）並根據 $0.5$ 的門檻值判定最終生死。

---

#### 🎯 開發歷程：這段程式碼背後的指令 (Prompts)

這是一個高度整合的請求，我們透過以下指令請 AI 處理繁瑣的數據對齊與預測邏輯：

> **指令：建立主角資料與預測**
> 「請幫我建立 Jack 和 Rose 兩筆資料，並對其進行與訓練資料一致的預處理（包含編碼與標準化），最後輸出他們的存活機率與預測結果。」

---

> 🎓 **給學生的學習小提醒**
>
> 在這個步驟中，最容易出錯的地方就是**「特徵順序」**。AI 的大腦是根據欄位的排列順序來認人的，如果 Jack 的年齡跟票價位置放反了，AI 就會給出完全錯誤的答案。因此，我們在程式碼中特別強調了 `original_X_cols` 的順序對齊。

````python
# 建立新的乘客資料 — 鐵達尼號的男女主角
new_passengers_data = [
    {'pclass': 3, 'sex': 'male',   'age': 23, 'sibsp': 1, 'parch': 0, 'fare': 5,   'embarked': 'S'},
    {'pclass': 1, 'sex': 'female', 'age': 20, 'sibsp': 1, 'parch': 0, 'fare': 100, 'embarked': 'S'}
]
new_passengers_df = pd.DataFrame(new_passengers_data)

print("原始新乘客資料:")
print(new_passengers_df)

# === 預處理新乘客資料,使其與訓練資料格式一致 ===
new_passengers_processed = new_passengers_df.copy()

# 處理 'sex' 欄位
new_passengers_processed['sex'] = new_passengers_processed['sex'].map({'female': 0, 'male': 1})

# 處理 'embarked' 欄位 (One-hot encoding)
embarked_dummies = pd.get_dummies(new_passengers_processed['embarked'], prefix='embarked')
new_passengers_processed = pd.concat([new_passengers_processed.drop('embarked', axis=1), embarked_dummies], axis=1)

# 確保所有 embarked 欄位都存在
for col in ['embarked_C', 'embarked_Q', 'embarked_S']:
    if col not in new_passengers_processed.columns:
        new_passengers_processed[col] = False
    new_passengers_processed[col] = new_passengers_processed[col].astype(int)

# 重新排列欄位順序,使其與訓練資料一致
original_X_cols = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare',
                   'embarked_C', 'embarked_Q', 'embarked_S']
new_passengers_processed = new_passengers_processed[original_X_cols]

# 對 'age' 和 'fare' 欄位進行 Min-Max 標準化(用之前訓練時 fit 過的 scaler)
new_passengers_processed[['age', 'fare']] = scaler.transform(new_passengers_processed[['age', 'fare']])

# 轉成 numpy 陣列、進行預測
new_passengers_array = new_passengers_processed.to_numpy(dtype=float)
predictions = model.predict(new_passengers_array)

print("\n=== 新乘客生存機率預測 ===")
print(f"🧑 Jack 的生存機率:{predictions[0][0]:.4f}")
print(f"👩 Rose 的生存機率:{predictions[1][0]:.4f}")

# 根據機率轉換為二元預測
predicted_labels = (predictions > 0.5).astype(int)
print("\n=== 新乘客生存預測 (0=死亡, 1=生存) ===")
print(f"🧑 Jack 的預測結果:{predicted_labels[0][0]}")
print(f"👩 Rose 的預測結果:{predicted_labels[1][0]}")
````

## 🔍 延伸戲：模型也會看走眼——找出「該活卻罹難」的人

這是整份教材中**最有教育意義的一段**。

目前為止，我們看到的都是 AI 準確預測的案例。但為了更進一步理解模型，我們要把全部 $1300$ 多位乘客（包含我們剛加入的 Jack 與 Rose）全部丟回給模型，讓它預測每個人的存活機率，然後提出一個核心疑問：

> **有沒有一些人，模型非常有信心他們會活（機率 $> 0.8$），但實際上他們卻在海難中罹難了？**

如果有，這代表即便模型再聰明，依然存在「意料之外」的盲點。在機器學習（ML）領域中，這種情況被稱為「**高信心錯誤 (High-confidence Errors)**」。

---

### 📋 接下來的程式碼會做三件事：

1.  **整合名單**：將 Jack、Rose 與原始的 $1300$ 多名乘客合併為一份完整名單。
2.  **批次預測**：呼叫模型，計算出這份名單中「每一位乘客」的生存機率。
3.  **逆向篩選**：找出那些「模型預測生還機率 $> 80\%$」但「實際結果卻是死亡 (survived=0)」的特殊個案。

---

#### 🎯 開發歷程：這段程式碼背後的指令 (Prompt)

這是一個涉及資料過濾與邏輯判斷的請求：

> **指令：找出高信心預測錯誤的乘客**
> 「請將預測後的存活機率加回原始資料表中。接著篩選出：實際上已經死亡（`survived=0`），但模型預測其生存機率卻大於 $0.8$ 的乘客名單，並印出他們的名字與各項特徵。」

---

> 🎓 **給學生的學習小提醒**
>
> 當你看到輸出的名單時，請不要覺得模型「壞掉了」。相反地，這些名單是我們優化模型最重要的線索。
>
> 每一位被 AI 誤判的人，背後可能都代表著一段數據無法捕捉的真實歷史（例如：一位頭等艙女性決定留在船上陪她的丈夫）。透過分析這些「例外」，我們才能真正理解數據科學與真實世界之間的鴻溝。

````python
# 建立包含 'name' 和 'survived' 的新乘客資料 DataFrame
new_passengers_with_details = pd.DataFrame({
    'pclass':   [3, 1],
    'survived': [0, 1],
    'name':     ['Jack', 'Rose'],
    'sex':      ['male', 'female'],
    'age':      [23, 20],
    'sibsp':    [1, 1],
    'parch':    [0, 0],
    'fare':     [5, 100],
    'embarked': ['S', 'S']
})

# 從原始 df 中選取需要的欄位
df_filtered = df[['pclass', 'survived', 'name', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']]

# 把 Jack、Rose 加入完整名單
df_combined = pd.concat([df_filtered, new_passengers_with_details], ignore_index=True)

# === 對 df_combined 做和訓練資料一樣的前處理 ===
all_feature_df = df_combined.copy()

all_feature_df['sex'] = all_feature_df['sex'].map({'female': 0, 'male': 1})

embarked_temp = all_feature_df['embarked']
all_feature_df = all_feature_df.drop(columns=['embarked', 'survived', 'name'])
embarked_dummies = pd.get_dummies(embarked_temp, prefix='embarked')
all_feature_df = pd.concat([all_feature_df, embarked_dummies], axis=1)

for col in ['embarked_C', 'embarked_Q', 'embarked_S']:
    if col not in all_feature_df.columns:
        all_feature_df[col] = False
    all_feature_df[col] = all_feature_df[col].astype(int)

original_X_cols = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare',
                   'embarked_C', 'embarked_Q', 'embarked_S']
all_feature_df = all_feature_df[original_X_cols]
all_feature_df[['age', 'fare']] = scaler.transform(all_feature_df[['age', 'fare']])

all_feature = all_feature_df.to_numpy(dtype=float)

# === 用模型預測所有人的存活機率,並加到 df_combined 成為新欄位 ===
probabilities = model.predict(all_feature)
df_combined['probability'] = probabilities

# === 篩選:實際上死了(survived=0),但模型很有把握會活(probability > 0.8)===
high_confidence_errors = df_combined[
    (df_combined['survived'] == 0) & (df_combined['probability'] > 0.8)
]

print(f"\n找到 {len(high_confidence_errors)} 位「死了卻被高度預測會活」的乘客:\n")
print(high_confidence_errors[['name', 'pclass', 'sex', 'age', 'fare', 'survived', 'probability']])
````
